```python
import pandas as pd

df = pd.read_csv('merged-dataset.csv', parse_dates=['start_date', 'end_date', 'inv_item_date'])
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df['yield_fact'] = df['yield_fact'].str.replace(',', '.').astype(float)
df['fertility_score'] = pd.to_numeric(df['fertility_score'], errors='coerce')
df['season'] = pd.to_numeric(df['season'], errors='coerce')
df['completed_work_area'] = pd.to_numeric(df['completed_work_area'], errors='coerce')
df['field_area'] = pd.to_numeric(df['field_area'], errors='coerce')
df['yield_plan'] = pd.to_numeric(df['yield_plan'], errors='coerce')
df['p2o5'] = pd.to_numeric(df['p2o5'], errors='coerce')
df['k2o'] = pd.to_numeric(df['k2o'], errors='coerce')
df['humus'] = pd.to_numeric(df['humus'], errors='coerce')
df['ph'] = pd.to_numeric(df['ph'], errors='coerce')
df['inv_item_weight'] = pd.to_numeric(df['inv_item_weight'], errors='coerce')
df['inv_item_dose'] = pd.to_numeric(df['inv_item_dose'], errors='coerce')
df['proc_depth'] = pd.to_numeric(df['proc_depth'], errors='coerce')
df['prod_purpose'] = pd.to_numeric(df['prod_purpose'], errors='coerce')
df['prod_weight'] = pd.to_numeric(df['prod_weight'], errors='coerce')
df['prod_dry_perc'] = pd.to_numeric(df['prod_dry_perc'], errors='coerce')
df['prod_humid'] = pd.to_numeric(df['prod_humid'], errors='coerce')
df['prod_impur'] = pd.to_numeric(df['prod_impur'], errors='coerce')
```
