```python
works = works[~works['work_subtype'].isin(['Вывоз', 'Уборка силос/карнаж', 'Сгребание в валки', 'Прессование', 'Подбор и обмолот',
                                           'Подбор (корм)', 'Подбор', 'Ворошение', 'Вывоз (тех.культуры)', 'Заготовка сенажа',
                                           'Уборка камней', 'Уборка трав'])]
len(works)
```

```python
grouped = works.groupby(['mech', 'work_type', 'work_subtype']).agg(amount_of_work=('work_subtype', 'count'), total_completed_work_area=('completed_work_area', 'sum')).reset_index()
grouped = grouped.sort_values(by=['mech', 'work_subtype', 'total_completed_work_area'], ascending=[True, True, False]).reset_index()

print(grouped.head(15))
```

```python
grouped = works.groupby(['mech', 'work_type', 'work_subtype']).agg(
    amount_of_work=('work_subtype', 'count'),
    total_completed_work_area=('completed_work_area', 'sum')
).reset_index()

grouped['specialization_index'] = grouped['amount_of_work'] * grouped['total_completed_work_area']

grouped['cumulative_sum'] = grouped.groupby('mech')['total_completed_work_area'].cumsum()
grouped['total_work_area_per_mech'] = grouped.groupby('mech')['total_completed_work_area'].transform('sum')
grouped['cumulative_percentage'] = grouped['cumulative_sum'] / grouped['total_work_area_per_mech']

grouped = grouped.sort_values(by=['mech', 'specialization_index'], ascending=[True, False])

print(grouped)
```

```python
works['start_date'] = pd.to_datetime(works['start_date'])
works = works.sort_values(by='start_date')
```

```python
works['work_number'] = works.groupby(['mech', 'work_subtype']).cumcount() + 1
```

```python
def count_unique_before(group):
    group['is_new_subtype'] = ~group['work_subtype'].duplicated()
    group['unique_works_before'] = group['is_new_subtype'].cumsum()
    return group

works = works.groupby('mech').apply(count_unique_before).reset_index(drop=True)
works.drop(columns=['is_new_subtype'], inplace=True)
```

```python
len(works)
```

```python
works['works_before'] = works.groupby('mech').cumcount() + 1
```

```python
works['metric'] = works['work_number'] / works['works_before']
```

```python
def calculate_cumsum_and_total(group):
    # Накопительная сумма для каждого work_subtype
    group['cumsum_subwork_area'] = group.groupby('work_subtype')['completed_work_area'].cumsum()
    
    # Общая сумма по work_subtype
    group['total_subwork_area'] = group.groupby('work_subtype')['completed_work_area'].transform('sum')
    
    return group

# Применяем функцию к каждой группе mech
works = works.groupby('mech').apply(calculate_cumsum_and_total).reset_index(drop=True)
works['total_area_for_mech'] = works.groupby('mech')['completed_work_area'].transform('sum')
works['cumsum_area'] = works.groupby('mech')['completed_work_area'].cumsum()

works['Metric_Area'] = works['cumsum_subwork_area'] / works['cumsum_area']
```

```python
import pandas as pd

# Пример данных (замените на ваш DataFrame)

# Функция для Target Encoding
def target_encode(data, categorical_cols, target_col):
    for col in categorical_cols:
        # Вычисляем среднее значение целевой переменной для каждой категории
        category_means = data.groupby(col)[target_col].transform('mean')
        # Добавляем новый столбец с закодированными значениями
        data[f'{col}_encoded'] = category_means
    return data

# Применение Target Encoding
print(works.columns)
categorical_cols = ['crop', 'crop_predecessor', 'work_type', 'work_subtype']
target_col = 'field_yield'
data_encoded = target_encode(works, categorical_cols, target_col)

data_encoded = data_encoded.drop(columns=categorical_cols)

df2 = data_encoded
```

```python
df2 = pd.get_dummies(works['region'])
for i in df2.columns:
    works[i] = df2[i]
df2 = pd.get_dummies(works['reproduction'])
for i in df2.columns:
    works[i] = df2[i]
df2 = pd.get_dummies(works['seed_origin'])
for i in df2.columns:
    works[i] = df2[i]
df2 = pd.get_dummies(works['technology'])
for i in df2.columns:
    works[i] = df2[i]
df2 = pd.get_dummies(works['land_type'])
for i in df2.columns:
    works[i] = df2[i]
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from statsmodels.compat import lzip
import statsmodels.stats.api as sms


df = works
df['same'] = df['crop'] != df['crop_predecessor']

df = df.drop(columns=['start_date', 
                      'end_date', 'field_yield', 'std', 'mean', 
                      'work_number', 'unique_works_before', 'works_before', 'cumsum_subwork_area',
                      'work_id', 'field', 'total_subwork_area', 'total_area_for_mech', 'cumsum_area',
                      'seed_origin', 'technology', 'prod_purpose', 'prod_weight', 'prod_dry_perc',
                      'prod_humid', 'prod_impur', 'work_type', 'work_subtype', 'mech', 
                      'region', 'land_type', 'crop', 'crop_predecessor',
                      'hybrid',	'reproduction',	'seed_origin', 'technology',
                      'main_tech', 'add_tech', 'completed_work_area', 'field_area', 'work_yield'])
df[df.columns] = df[df.columns].apply(pd.to_numeric, errors='coerce')

df = df.fillna(0)  # works

numeric_columns = df.select_dtypes(include=['int', 'float']).columns
scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
X = df.drop(columns=['yield_z', 'metric'])
y = df['yield_z']

model = sm.OLS(y.astype('float'), sm.add_constant(X.astype('float'))).fit()
# test_result = sms.het_breuschpagan(model.resid, model.model.exog)

print(model.summary())
```
