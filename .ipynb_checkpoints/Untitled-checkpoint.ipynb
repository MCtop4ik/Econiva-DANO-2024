```python
import pandas as pd

df = pd.read_csv('works_finals.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
unique_works_by_mech = df.groupby('mech')['work_subtype'].nunique().reset_index()
unique_works_by_mech.rename(columns={'work_subtype': 'unique_works'}, inplace=True)

total_works_by_mech = df.groupby('mech')['work_subtype'].count().reset_index()
total_works_by_mech.rename(columns={'work_subtype': 'total_works'}, inplace=True)

merged_df = pd.merge(unique_works_by_mech, total_works_by_mech, on='mech')
merged_df['ratio_unique_to_total'] = merged_df['unique_works'] / merged_df['total_works']
```

```python
df = pd.merge(df, merged_df, on='mech')
```

```python
df.to_csv('with-controll-variable.csv')
```
