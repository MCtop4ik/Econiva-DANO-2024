```python
import pandas as pd

works = pd.read_csv('ekoniva-dataset/works_new.csv')
fields = pd.read_csv('ekoniva-dataset/fields_new.csv')
```

```python
df = works.merge(fields, 'left', on=['field', 'season'])
df = df.rename(columns={
    'area_x': 'completed_work_area',
    'area_y': 'field_area'
})
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Пример данных (ваш DataFrame)
# df = pd.read_csv('your_data.csv')  # Загрузите ваш DataFrame

# Группировка данных: подсчет количества выполнений для каждого специалиста (mech) по каждой подработе
subtype_counts = df.groupby(['work_subtype', 'mech']).size().reset_index(name='count')

# Используем FacetGrid для построения графиков для каждой подработы
g = sns.FacetGrid(subtype_counts, col='work_subtype', col_wrap=3, height=4, sharex=False, sharey=False)
g.map(sns.histplot, 'count', bins=10, kde=True, color='blue')

# Настройка графика
g.set_titles(col_template='{col_name}', fontsize=14)
# g.set_xlabels('Количество выполнений', fontsize=12)
# g.set_ylabels('Количество специалистов', fontsize=12)

# Отображение графика
plt.tight_layout()
plt.show()
```

```python
# Группировка по человеку, типу и подтипу работ, суммирование объёма
grouped = df.groupby(['person', 'work_type', 'work_subtype'])['completed_work_area'].sum().reset_index(name='total_completed_work_area')

# Сортировка по общему объёму работ
grouped = grouped.sort_values(by=['person', 'total_completed_work_area'], ascending=[True, False])

# Вывод результата
print(grouped)
```
