```python
import pandas as pd
```

```python
works = pd.read_csv('ekoniva-dataset/works_new.csv')
fields = pd.read_csv('ekoniva-dataset/fields_new.csv')
value = pd.read_csv('ekoniva-dataset/value_new.csv', parse_dates=['inv_item_date'], dayfirst=True)
```

```python
df_intermediate = works.merge(fields, 'left', on=['field', 'season'])
```

```python
df_intermediate.columns
```

```python
df = df_intermediate.merge(value, "left", on=["field", "work_id"])
```

```python
df = df.rename(columns={
    'area_x': 'completed_work_area',
    'area_y': 'field_area'
})
```
